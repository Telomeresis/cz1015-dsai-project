```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
# Import into pandas dataframe
data16 = pd.read_excel('16raw_datawig_imputed.xlsx')
```

```python
print("Data type : ", type(data16))
print("Data dims : ", data16.shape)
print()
print(data16.dtypes)
```

```python
# Uses values provided from World Happiness Report
predictors = ["Log GDP per capita (PPP)", "Social Support", "Log Healthy life expectancy at birth", "Freedom to make life choices", "Generosity", "Perceptions of corruption", "Happiness"]

# Extract only the numeric data variables
numDF = pd.DataFrame(data16[predictors])

# Summary Statistics for all Variables
numDF.describe()
```

```python
# Draw the Boxplots of all variables
f, axes = plt.subplots(1, 1, figsize=(24, 12))
sb.boxplot(data = numDF, orient = "h", color="grey")
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(7, 2, figsize=(12, 34))

count = 0
for var in numDF:
    sb.distplot(numDF[var], ax = axes[count,0], color="grey")
    sb.violinplot(numDF[var], ax = axes[count,1], color="red")
    count += 1
```

```python
# Calculate the complete  correlation matrix
numDF.corr()
```

```python
# Heatmap of the Correlation Matrix
f, axes = plt.subplots(1, 1, figsize=(12, 8))
sb.heatmap(numDF.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = numDF)
```
